```python
import numpy as np
import scipy as sp
from scipy import integrate
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation as anim
import random

%matplotlib notebook
```

```python
def vdp(state, t, ep):
    x1, x2 = state
    return [x2, -ep * (x1 ** 2.0 - 1.0) * x2 - x1]
```

```python
for epsilon in [0.01, 0.2, 1, 5.0, 100.0]:
    plt.figure(figsize=(15, 15))

    for i in range(10):

        rng = 5
        x, y = (random.random()-0.5)*rng, (random.random()-0.5)*rng

        time_rng = 50
        t = np.linspace(0, time_rng, 100*time_rng)
        ts = np.linspace(0.0, 100.0, 50000)
        pth = integrate.odeint(vdp, (x, y), ts, (epsilon,))
        plt.plot(pth[:,0], pth[:,1])
    
    plt.title(f"$ \epsilon $ = {epsilon}")
    plt.show()
```
